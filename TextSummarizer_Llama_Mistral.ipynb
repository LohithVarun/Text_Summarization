```python
from unsloth import FastLanguageModel
from unsloth.chat_templates import get_chat_template
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import gc
from PyPDF2 import PdfReader
from google.colab import files
import io
import os
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import nltk



def upload_and_read_pdf():
    """Upload and read PDF file"""
    print("Please upload your research paper PDF file...")
    uploaded = files.upload()

    if not uploaded:
        raise Exception("No file was uploaded")

    file_name = list(uploaded.keys())[0]
    pdf_content = uploaded[file_name]

    # Read PDF content
    pdf_reader = PdfReader(io.BytesIO(pdf_content))
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text()

    # Basic text cleaning
    text = text.replace('\n', ' ').replace('\r', ' ')
    text = ' '.join(text.split())

    print(f"\nSuccessfully read PDF file: {file_name}")
    print(f"Total characters: {len(text)}")
    return text

def initialize_model(model_name):
    """Initialize model using Unsloth's FastLanguageModel"""
    try:
        print(f"Loading {model_name} with Unsloth...")
        # Initialize model using unsloth
        model, tokenizer = FastLanguageModel.from_pretrained(
            model_name=model_name,
            max_seq_length=2048*3, # Adjusted max_seq_length, can be tuned
            dtype=None,  # Auto detect
            load_in_4bit=True
        )

        # For inference - important for Unsloth models
        FastLanguageModel.for_inference(model)

        # Set up chat template - if needed, adjust template name
        tokenizer = get_chat_template(
            tokenizer,
            chat_template="llama-3.1", # or "mistral" if using mistral model and template, auto-detected now
        )
        return model, tokenizer
    except Exception as e:
        print(f"Error initializing model {model_name} with Unsloth: {str(e)}")
        return None, None

def chunk_text(text, chunk_size=1000):
    """Split text into smaller chunks"""
    words = text.split()
    chunks = []
    current_chunk = []
    current_length = 0

    for word in words:
        current_length += len(word) + 1  # +1 for space
        if current_length > chunk_size:
            chunks.append(' '.join(current_chunk))
            current_chunk = [word]
            current_length = len(word)
        else:
            current_chunk.append(word)

    if current_chunk:
        chunks.append(' '.join(current_chunk))

    return chunks

def generate_summary(model, tokenizer, text):
    """Generate summary without beam search using Unsloth model"""
    try:
        # Prepend instruction for better summaries
        prompt = f"Please summarize the following text concisely:\n\n{text}\n\nSummary:"

        messages = [{"role": "user", "content": prompt}]

        inputs = tokenizer.apply_chat_template(
            messages,
            tokenize=True,
            add_generation_prompt=True,
            return_tensors="pt"
        ).to(model.device)


        with torch.no_grad():
            outputs = model.generate(
                input_ids=inputs,
                max_new_tokens=100,
                do_sample=True,
                temperature=0.7,
                top_p=0.9,
                top_k=50,
                repetition_penalty=1.2,
                pad_token_id=tokenizer.eos_token_id, # Might not be needed for Unsloth
                use_cache=True # Recommended for faster generation
            )

        # Extract only the summary part (after our prompt)
        full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
        summary = full_output.split("Summary:")[-1].strip()

        del outputs
        torch.cuda.empty_cache()
        return summary
    except Exception as e:
        print(f"Error generating summary: {str(e)}")
        return ""

def calculate_relevance_embedding_similarity(source_text, summary_text):
    """Calculates relevance using cosine similarity of sentence embeddings."""
    model = SentenceTransformer('all-MiniLM-L6-v2') # Choose a suitable sentence embedding model

    source_embeddings = model.encode(source_text, convert_to_tensor=True)
    summary_embeddings = model.encode(summary_text, convert_to_tensor=True)

    # For simplicity, averaging embeddings of sentences in source and summary.
    # For better results, you might want to process sentence by sentence and then aggregate.
    source_avg_embedding = torch.mean(source_embeddings, dim=0).cpu().numpy().reshape(1, -1) # Reshape for cosine_similarity
    summary_avg_embedding = torch.mean(summary_embeddings, dim=0).cpu().numpy().reshape(1, -1) # Reshape for cosine_similarity

    similarity_score = cosine_similarity(source_avg_embedding, summary_avg_embedding)[0][0]
    return similarity_score



def calculate_factcc_faithfulness(source_text, summary_text):
    """Calculates FactCC faithfulness score using NLI, attempting text_classification pipeline."""

    if not summary_text.strip():  # Handle empty summary case
        return 1.0 # Or 0.0, depending on how you want to score empty summaries. 1.0 is more lenient.

    nli_pipeline = pipeline("text-classification", model="roberta-large-mnli", device=0) # Using text-classification pipeline, explicitly set device

    source_sentences = nltk.sent_tokenize(source_text)
    summary_sentences = nltk.sent_tokenize(summary_text)

    entailment_count = 0
    for summary_sentence in summary_sentences:
        premise = source_text  # Consider the entire source chunk as premise for each summary sentence
        hypothesis = summary_sentence

        try: # Add try-except for potential errors in pipeline execution
            nli_result = nli_pipeline(f"{hypothesis} </s></s> {premise}") # Concatenate hypothesis and premise with separators
            predicted_label = nli_result[0]['label']

            if predicted_label == 'LABEL_0' or predicted_label == 'entailment': # Check for both LABEL_0 and 'entailment' (different versions/outputs)
                entailment_count += 1
        except Exception as e:
            print(f"Warning: Error during NLI inference for sentence: '{summary_sentence}'. Skipping. Error: {e}")
            continue # Skip to the next sentence in case of error

    faithfulness_score = entailment_count / len(summary_sentences) if summary_sentences else 1.0 # Avoid division by zero
    return faithfulness_score

def evaluate_model(model, tokenizer, text, num_chunks=5):
    """Evaluate model with memory management"""
    scores = {
        'faithfulness': [],
        'relevance': []
    }

    # Split text into chunks
    chunks = chunk_text(text)
    print(f"\nProcessing {len(chunks)} text chunks for evaluation...")

    # Limit chunks to evaluate to save time
    chunks_to_process = chunks[:min(num_chunks, len(chunks))]

    for chunk in tqdm(chunks_to_process, desc="Evaluating chunks"):
        summary = generate_summary(model, tokenizer, chunk)
        if summary:
            try:
                # Calculate FactCC faithfulness score
                faithfulness_score = calculate_factcc_faithfulness(chunk, summary)
                scores['faithfulness'].append(faithfulness_score)

                # Calculate relevance using embedding similarity (already implemented)
                relevance_score = calculate_relevance_embedding_similarity(chunk, summary)
                scores['relevance'].append(relevance_score)

                gc.collect()
                torch.cuda.empty_cache()

            except Exception as e:
                print(f"Error during metric evaluation: {str(e)}")
                continue

    # Calculate average scores
    return {k: np.mean(v) if v else 0 for k, v in scores.items()}

def plot_metrics(model1_scores, model2_scores, model1_name, model2_name):
    """Plot comparison graphs interactively"""
    metrics = ['faithfulness', 'relevance'] # Adjusted metrics list

    # Plot first model metrics
    plt.figure(figsize=(8, 5))
    plt.bar(metrics, [model1_scores['faithfulness'], model1_scores['relevance']], color='blue') # Adjusted values
    plt.title(f'{model1_name} Model Metrics')
    plt.ylim(0, 1)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Plot second model metrics
    plt.figure(figsize=(8, 5))
    plt.bar(metrics, [model2_scores['faithfulness'], model2_scores['relevance']], color='green') # Adjusted values
    plt.title(f'{model2_name} Model Metrics')
    plt.ylim(0, 1)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Comparative plot
    plt.figure(figsize=(10, 5))
    x = np.arange(len(metrics))
    width = 0.35

    plt.bar(x - width/2, [model1_scores['faithfulness'], model1_scores['relevance']], width, label=model1_name, color='blue') # Adjusted values
    plt.bar(x + width/2, [model2_scores['faithfulness'], model2_scores['relevance']], width, label=model2_name, color='green') # Adjusted values

    plt.xlabel('Metrics')
    plt.ylabel('Scores')
    plt.title(f'{model1_name} vs {model2_name} Performance Comparison')
    plt.xticks(x, metrics, rotation=45)
    plt.legend()
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.show()

def main():
    try:
        # Upload and read PDF
        print("Welcome to Research Paper Summarization Analyzer")
        paper_text = upload_and_read_pdf()

        if not paper_text:
            print("No text could be extracted from the PDF. Exiting...")
            return

        print("\nInitializing models and generating summaries. This may take a few minutes...")

        # Use open-access Llama and Mistral alternatives that don't require authentication
        model1_name = "unsloth/Llama-3.2-3B-Instruct"  # Unsloth Llama 3.2 model
        model2_name = "unsloth/mistral-7b-instruct-v0.3-bnb-4bit"  # Unsloth Mistral model (4bit)

        # Fallback models - using standard HF models as fallback, can be adjusted
        fallback_model1 = "facebook/opt-1.3b"
        fallback_model2 = "bigscience/bloom-1b7"

        # Process first model (Unsloth Llama)
        print(f"\nProcessing with {model1_name} model...")
        model1, tokenizer1 = initialize_model(model1_name) # Initialize using Unsloth
        if model1 is None:
            print(f"Failed to initialize {model1_name} model. Trying fallback...")
            model1_name = fallback_model1
            model1, tokenizer1 = initialize_model(model1_name) # Fallback to standard HF init
            if model1 is None:
                print(f"Failed to initialize fallback model {model1_name}. Skipping...")
                model1_scores = {'faithfulness': 0, 'relevance': 0} # Adjusted scores dict
            else:
                print(f"Evaluating fallback model {model1_name}...")
                model1_scores = evaluate_model(model1, tokenizer1, paper_text)
        else:
            print(f"Evaluating {model1_name}...")
            model1_scores = evaluate_model(model1, tokenizer1, paper_text)

        del model1, tokenizer1
        gc.collect()
        torch.cuda.empty_cache()

        # Process second model (Mistral) - Using Unsloth Mistral model as requested
        print(f"\nProcessing with {model2_name} model...")
        model2, tokenizer2 = initialize_model(model2_name) # Try Unsloth init for Mistral
        if model2 is None:
            print(f"Failed to initialize {model2_name} model with Unsloth. Trying standard fallback...")
            model2_name = fallback_model2 # Using a different fallback for model2
            model2, tokenizer2 = initialize_model(model2_name) # Fallback to standard HF init for fallback model
            if model2 is None:
                print(f"Failed to initialize fallback model {model2_name}. Skipping...")
                model2_scores = {'faithfulness': 0, 'relevance': 0} # Adjusted scores dict
            else:
                print(f"Evaluating fallback model {model2_name}...")
                model2_scores = evaluate_model(model2, tokenizer2, paper_text)
        else:
            print(f"Evaluating {model2_name}...")
            model2_scores = evaluate_model(model2, tokenizer2, paper_text)


        del model2, tokenizer2
        gc.collect()
        torch.cuda.empty_cache()

        # Display results
        print("\nEvaluation Results:")
        print(f"\n{model1_name} Scores:")
        # Removed hallucination from print output
        print(f"faithfulness: {model1_scores['faithfulness']:.3f}")
        print(f"relevance: {model1_scores['relevance']:.3f}")

        print(f"\n{model2_name} Scores:")
        # Removed hallucination from print output
        print(f"faithfulness: {model2_scores['faithfulness']:.3f}")
        print(f"relevance: {model2_scores['relevance']:.3f}")

        # Display plots
        print("\nGenerating comparison plots...")
        model1_display_name = model1_name.split('/')[-1]
        model2_display_name = model2_name.split('/')[-1]
        plot_metrics(model1_scores, model2_scores, model1_display_name, model2_display_name)

    except Exception as e:
        print(f"Error in main execution: {str(e)}")

if __name__ == "__main__":
    main()
```
